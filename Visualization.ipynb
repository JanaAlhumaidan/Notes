1. Scatterplot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

2. Bar

```python
nut = {"Almond": 10,
        "peanut":8,
        "cashew": 12}
fig, ax = plt.subplots()
ax.bar(nut.keys(), nut.values())
ax.set(title= 'Nuts', ylabel='Price')
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut.keys()), list(nut.values()))
ax.set(title= 'Nuts', xlabel='Price')
```

3. Histograms

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

4. Subplot

```python
# option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, 
                                               ncols=2, 
                                               figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut.keys(), nut.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots( nrows=2, 
                        ncols=2, 
                        figsize=(10,5))

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut.keys(), nut.values());
ax[1,1].hist(np.random.randn(1000));
```

8. pie

```python
import pandas as pd
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(1, 100, size=len(categories))
data = pd.DataFrame({'Categories': categories, 'Values': values})

# Create a pie chart
fig = px.pie(data, values='Values', names='Categories', title='Pie Chart')
fig
```
